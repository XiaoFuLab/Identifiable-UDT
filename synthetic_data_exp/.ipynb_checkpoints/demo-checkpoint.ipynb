# DIMENSION with I=3

```python
'''
Import packages and config
'''

import argparse
import torch
import numpy as np
from data_loader import get_all_loaders_gaussian
import yaml 
from model import Generator, Critic
from trainer import Trainer
import os
import matplotlib.pyplot as plt

torch.backends.cudnn.enabled = True
torch.backends.cudnn.benchmark = True

seed = 1000
np.random.seed(seed)
torch.manual_seed(seed)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

def str2bool(v):
    return v.lower() in ('true')

parser = argparse.ArgumentParser()
parser.add_argument('--config_file', type=str, default='configs/gaussian.yaml', help='Path to the config file.')

args = parser.parse_args([])
# load config file
with open(args.config_file) as f:
    config = yaml.safe_load(f)
    
num_components = 3

# use num_conditionals = 1 for CycleGAN-Loss
num_conditionals = 3

```

```python
'''
Generate data and construct data loaders
'''

train_loader, test_loader = get_all_loaders_gaussian(dtype=torch.float32, 
                                                                batch_size=config['batch_size'], 
                                                                num_train_samples=config['num_train_samples'], 
                                                                num_test_samples=config['num_test_samples'], 
                                                                shuffle=True,
                                                                num_components=num_components,
                                                                num_conditionals=num_conditionals,
                                                                dim=2,
                                                                variance=0.3,
                                                                save_filename=os.path.join(config['data_path'], config['save_data_file']))
train_loader_iter = iter(train_loader)

print('Consructing test set..')
test_x1, test_x2 = [], []
test_iters = config['test_size']//config['batch_size']
for it, (x1, _, x2, _) in enumerate(test_loader):
    test_x1.append(x1)
    test_x2.append(x2)
    if it>test_iters:
        break

test_x1, test_x2 = torch.cat(test_x1, dim=0).to(device), torch.cat(test_x2, dim=0).to(device)
print('Test set constructed.')
```

```python
'''
Create and train model
'''

# load model
trainer = Trainer(config=config, train_loader=train_loader, device=device, num_conditionals=num_conditionals)

# train model
for it in range(config['train_iters']):
    try:
        x_1, l_1, x_2, l_2 = next(train_loader_iter)
    except:
        train_loader_iter = iter(train_loader)
        x_1, l_1, x_2, l_2 = next(train_loader_iter)

    x_1, l_1, x_2, l_2 = x_1.to(trainer.device), l_1.to(trainer.device), x_2.to(trainer.device), l_2.to(trainer.device)
    
    trainer.dis_update(x_1, l_1, x_2, l_2)
    trainer.gen_update(x_1, l_1, x_2, l_2)
    
    if (it+1) % config['console_log_interval'] == 0:
        trainer.log_err_console(it)
        
    if it % config['plot_interval'] ==0:
        xt_1, xt_121, xt_12, xt_2, xt_212, xt_21 = trainer.sample(test_x1, test_x2)
        trainer.translation_err(xt_1, xt_2, xt_12, xt_21)
        trainer.plot_scatter(xt_1, xt_2, xt_12, xt_21)
        plt.show()
```
